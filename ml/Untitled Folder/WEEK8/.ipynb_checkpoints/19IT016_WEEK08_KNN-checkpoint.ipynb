## Iris Dataset

**KNN implemetation using sklearn library**

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.neighbors import KNeighborsClassifier
```

# Dataset

**Iris dataset contains five columns such as Petal Length, Petal Width, Sepal Length, Sepal Width and Species Type. Iris is a flowering plant, the researchers have measured various features of the different iris flowers and recorded digitally.Based on that parameters going to define class**

```python
df = pd.read_csv("iris.csv")
df.head()
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
print(x)
print(y)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=2)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
x_test
```

```python
y_test
```

Because the KNN classifier predicts the class of given test observation by identifying observation nearest to it,the scale of variable matters.Any variable that on large scale will have a much larger effect on distance between the observations, and hence on knn classifier variables that are on small scale.

```python
scaler = StandardScaler()
scaler.fit(x_train)
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
knn =  KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train,y_train)
```

```python
y_pred = knn.predict(x_test)
```

```python
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

# Can we use KNN for regression also? Why / Why not?

**KNN algorithms can be used for both classification and regression problems.**

**Why not ?**
The k-nearest neighbors (KNN) algorithm is a simple, supervised machine learning algorithm that can be used to solve both classification and regression problems. It's easy to implement and understand, but has a major drawback of becoming significantly slower as the size of that data in use grows.
